Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a popular Python library used for creating static, interactive, and animated visualizations in a wide range of formats. It provides a robust set of tools for generating various types of plots and charts. The library is particularly well-suited for data visualization and scientific plotting due to its flexibility and ease of use.

Key features and reasons for using Matplotlib:

Flexible Plotting: Matplotlib allows you to create a wide variety of plots and customize them to suit your specific needs.

Data Visualization: It is extensively used for visualizing data to gain insights and communicate findings effectively.

Publication Quality: Matplotlib produces high-quality plots suitable for academic publications, presentations, and reports.

Interactive Visualization: It supports interactive features, which are useful for exploring data and adjusting visualizations on-the-fly.

Compatibility: Matplotlib can be used with various GUI toolkits, web applications, and different Python environments.

Line Plot: A basic plot that displays data points connected by straight lines, commonly used to show trends over time or a continuous sequence of data.

Scatter Plot: It represents individual data points as dots in a two-dimensional space, suitable for visualizing the relationship between two numerical variables.

Bar Plot: A plot where rectangular bars represent data values, often used to compare categorical data or display frequency distributions.

Histogram: This plot displays the distribution of a single numerical variable and helps identify the underlying data distribution.

Pie Chart: A circular chart divided into sectors, representing data proportions as slices of the pie, useful for showing parts of a whole.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

A scatter plot is a type of data visualization that represents individual data points as dots in a two-dimensional space. It is commonly used to visualize the relationship between two numerical variables and to identify patterns or correlations in the data.

```python
import numpy  as np
import matplotlib .pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y)
```

```python
plt.title("scatter plt")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

The subplot() function in Matplotlib is used to create multiple plots within the same figure, allowing you to arrange them in a grid-like fashion. This is particularly useful when you want to visualize multiple datasets or compare different p|lots side by side. The subplot() function takes three arguments: the number of rows, the number of columns, and the index of the subplot you want to create.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid of subplots and plot each line on a separate subplot
plt.subplot(2, 2, 1)
plt.plot(x1, y1, label='Line 1')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot 1')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(x2, y2, label='Line 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot 2')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(x3, y3, label='Line 3')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot 3')
plt.legend()
plt.subplot(2, 2, 4)
plt.plot(x4, y4, label='Line 4')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot 4')
plt.legend()

# Adjust the layout to prevent overlapping of titles and labels
plt.tight_layout()

# Show the plot
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a bar plot
plt.bar(company, profit)
plt.xlabel('Company')
plt.ylabel('Profit (in millions)')
plt.title('Company Profits')
plt.show()

# Create a horizontal bar plot
plt.barh(company, profit)
plt.xlabel('Profit (in millions)')
plt.ylabel('Company')
plt.title('Company Profits (Horizontal)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for box plots
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1,box2]
plt.boxplot(data,  labels=['box 1','box 2'])
plt.xlabel('box')
plt.title('box plot')
plt.show()
```
