Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would
use each type of test.

: T-test and Z-test are both statistical hypothesis tests used to make inferences about population parameters based on sample data, but they are typically used under different circumstances due to differences in assumptions. Here's a summary of the key differences and an example scenario for each test:

T-Test:

Assumption: The t-test assumes that the population follows a normal distribution, and it is best used when the sample size is small (typically less than 30) or when the population standard deviation is unknown.
Example Scenario: Suppose you want to determine if there is a significant difference in the mean exam scores between two different classes (Class A and Class B) in a small school. You have sample data from each class (e.g., 15 students in each class), and you do not know the population standard deviation. In this case, you would use a two-sample t-test.
Z-Test:

Assumption: The z-test assumes that the population follows a normal distribution, and it is appropriate when the sample size is large (typically greater than 30) or when the population standard deviation is known.
Example Scenario: Let's say you are a quality control manager at a factory, and you want to test if a new manufacturing process has significantly improved the mean weight of a product. You have a large sample size (e.g., 1000 products) from the new process, and you also know the population standard deviation based on historical data. In this case, you would use a z-test.

Q2: Differentiate between one-tailed and two-tailed tests.

One-tailed and two-tailed tests are two types of statistical hypothesis tests used to determine the significance of observed data. They differ in terms of the directionality of the hypotheses and the way they assess significance. Here's a differentiation between them:

One-Tailed Test:

Directionality: In a one-tailed test, you are testing a specific directional hypothesis. You are interested in whether the sample data is significantly different from a population parameter in one particular direction (either greater than or less than).
Hypotheses:
Null Hypothesis (H0): It typically states that there is no significant difference or no effect.
Alternative Hypothesis (Ha): It specifies the direction of the difference or effect you are interested in

Critical Region: The critical region, where you compare your sample statistic to a critical value, is located entirely on one side of the probability distribution (either to the right or left).
Use Case: One-tailed tests are often used when you have a specific hypothesis about the direction of the effect and are only interested in deviations in that one direction. For example, testing if a new drug has a better curing rate than an existing one

Two-Tailed Test:

Directionality: In a two-tailed test, you are testing a non-directional hypothesis. You are interested in whether the sample data is significantly different from a population parameter in either direction (greater than or less than).
Hypotheses:
Null Hypothesis (H0): It typically states that there is no significant difference or no effect.
Alternative Hypothesis (Ha): It doesn't specify a particular direction but simply states that there is a significant difference

Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.

A confidence interval is a range of values that is used to estimate an unknown population parameter, such as the mean, median, variance, or proportion, with a certain level of confidence. It quantifies the uncertainty associated with estimating the parameter from a sample of data. In other words, a confidence interval provides a range of values within which we believe the true population parameter is likely to fall, along with a specified level of confidence.

The formula for calculating a confidence interval for a population parameter typically involves the following components:

Point Estimate: This is a sample statistic that serves as our best estimate of the population parameter. For example, if you want to estimate the population mean 
μ), the sample mean (
x) is your point estimate.

Margin of Error: This represents the range of values added to and subtracted from the point estimate to create the interval. It is calculated based on the standard error of the point estimate and the desired level of confidence.

Level of Confidence: This is a percentage that indicates how confident you are that the true population parameter falls within the calculated interval. Common levels of confidence include 90%, 95%, and 99%.
