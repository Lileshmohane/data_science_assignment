Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.

```python
playstore_data= df
```

```python
playstore_data = playstore_data[playstore_data['Price'] != 'Everyone']

# Convert the 'Price' column to numeric
playstore_data['Price'] = pd.to_numeric(playstore_data['Price'].str.replace('$', ''))

# Create a bar chart to compare average prices by app type
plt.figure(figsize=(10, 6))
average_prices_by_type = playstore_data.groupby('Type')['Price'].mean().reset_index()
plt.bar(average_prices_by_type['Type'], average_prices_by_type['Price'], color='skyblue')
plt.title('Average Prices by App Type')
plt.xlabel('App Type')
plt.ylabel('Average Price')
plt.show()
```

Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps
with the highest number of installs.

```python
playstore_data['Installs'] = playstore_data['Installs'].str.replace(',', '')
playstore_data['Installs'] = playstore_data['Installs'].str.replace('+', '')
playstore_data['Installs'] = pd.to_numeric(playstore_data['Installs'], errors='coerce')

# Create a frequency table for app installations
install_freq_table = playstore_data.groupby('App')['Installs'].sum().reset_index()

# Sort the frequency table to get the top 10 most popular apps
top_10_apps = install_freq_table.sort_values(by='Installs', ascending=False).head(10)

# Display the top 10 most popular apps
print("Top 10 Most Popular Apps:")
print(top_10_apps[['App', 'Installs']])
```

```python
# Aggregate data by app category
category_metrics = playstore_data.groupby('Category').agg({
    'Installs': 'sum',
    'Rating': 'mean',
    'Reviews': 'sum'
}).reset_index()

# Rank categories based on chosen popularity metrics
category_metrics['PopularityRank'] = category_metrics['Installs'].rank(ascending=False)

# Display top categories by popularity
top_categories = category_metrics.sort_values(by='PopularityRank').head(10)
print("Top 10 App Categories by Popularity:")
print(top_categories[['Category', 'Installs', 'Rating', 'Reviews']])

# Create visualizations (e.g., bar charts, scatter plots) to present the findings
plt.figure(figsize=(12, 6))
sns.barplot(x='Category', y='Installs', data=top_categories, palette='viridis')
plt.title('Top 10 App Categories by Installs')
plt.xlabel('Category')
plt.ylabel('Total Installs')
plt.xticks(rotation=45, ha='right')
plt.show()
```

Q16. A mobile app development company wants to analyze the Google Playstore dataset to identify the
most successful app developers. What features would you analyze to make recommendations to the
company, and what data visualizations would you use to present your findings?

```python
playstore_data.head(2)
```

```python
developer_metrics = playstore_data.groupby('Developer').agg({
    'Installs': 'sum',
    'Rating': 'mean',
    'Reviews': 'sum',
    'App': 'count'
}).reset_index()

# Rank developers based on chosen success metrics
developer_metrics['SuccessRank'] = developer_metrics['Installs'].rank(ascending=False)

# Display top developers by success metrics
top_developers = developer_metrics.sort_values(by='SuccessRank').head(10)
print("Top 10 App Developers by Success Metrics:")
print(top_developers[['Developer', 'Installs', 'Rating', 'Reviews', 'App']])

# Create visualizations
plt.figure(figsize=(16, 12))
plt.subplot(3, 2, 1)
sns.barplot(x='Developer', y='Installs', data=top_developers, palette='viridis')
plt.title('Top 10 Developers by Total Installs')

plt.subplot(3, 2, 2)
sns.barplot(x='Developer', y='Rating', data=top_developers, palette='viridis')
plt.title('Top 10 Developers by Average Rating')

plt.subplot(3, 2, 3)
sns.barplot(x='Developer', y='Reviews', data=top_developers, palette='viridis')
plt.title('Top 10 Developers by Total Reviews')

plt.subplot(3, 2, 4)
sns.barplot(x='Developer', y='App', data=top_developers, palette='viridis')
plt.title('Top 10 Developers by App Count')

plt.subplot(3, 2, 5)
sns.scatterplot(x='Installs', y='Rating', data=developer_metrics, hue='Developer', palette='viridis')
plt.title('Installs vs. Average Rating')

plt.subplot(3, 2, 6)
sns.scatterplot(x='Installs', y='Reviews', size='Reviews', data=developer_metrics, hue='Developer', palette='viridis')
plt.title('Installs vs. Number of Reviews')

plt.tight_layout()
plt.show()
```

```python
playstore_data.head(2)
```

```python


# Convert 'Date' to datetime format

playstore_data['Date']=playstore_data['Last Updated']
playstore_data['Date'] = pd.to_datetime(playstore_data['Date'])
# Extract month, day, and hour features
playstore_data['Month'] = playstore_data['Date'].dt.month
playstore_data['DayOfWeek'] = playstore_data['Date'].dt.day_name()
playstore_data['Hour'] = playstore_data['Date'].dt.hour

# Create visualizations
plt.figure(figsize=(16, 12))

plt.subplot(3, 2, 1)
sns.lineplot(x='Date', y='Installs', data=playstore_data, estimator='sum', ci=None)
plt.title('Monthly Installs Over Time')

plt.subplot(3, 2, 2)
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
sns.heatmap(playstore_data.groupby(['Month', 'DayOfWeek'])['Installs'].sum().unstack(), cmap='viridis', annot=True, fmt="d", linewidths=.5, cbar_kws={'label': 'Total Installs'})
plt.title('Day of the Week Heatmap')

plt.subplot(3, 2, 3)
sns.lineplot(x='Month', y='Installs', data=playstore_data, estimator='sum', ci=None)
plt.title('Seasonal Variation in Installs')

plt.subplot(3, 2, 4)
sns.boxplot(x='Hour', y='Installs', data=playstore_data)
plt.title('Hourly Trends in Installs')

plt.subplot(3, 2, 5)
sns.boxplot(x='Month', y='Installs', data=playstore_data)
plt.title('Monthly Variation in Installs')

plt.subplot(3, 2, 6)
sns.boxplot(x='DayOfWeek', y='Installs', order=day_order, data=playstore_data)
plt.title('Day of the Week Variation in Installs')

plt.tight_layout()
plt.show()
```
